```python
def table1(a):
    driver=webdriver.Firefox()
    url=f'https://www.1mg.com/categories/homeopathy-57?filter=true&discount=10%2C20%2C30&pageNumber={a}'
    driver.get(url)
    soup=BeautifulSoup(driver.page_source,'html.parser')
    
    for i in range(len(soup.find_all('div',{'class':'col-md-3 col-sm-4 col-xs-6 style__container___1TL2R'}))):
        medicine = soup.find_all('div',{'class':'col-md-3 col-sm-4 col-xs-6 style__container___1TL2R'})[i]
        Name = medicine.find_all('div',{'class':'style__pro-title___2QwJy'})[0].text
        size= medicine.find_all('div',{'class':'style__pack-size___2JQG7'})[0].text
        MRP= medicine.find_all('span',{'class':'style__discount-price___25Bya'})[0].text
        price= medicine.find_all('div',{'class':'style__price-tag___cOxYc'})[0].text
        url = medicine.find_all('a',{'class':'style__product-link___UB_67'})[0].get('href')
    
    
    
        medicine_dict = {
            'Name' : Name,
            'size' : size,
            'MRP' : MRP,
            'price' : price,
            'url' : url

        }

        medicine_list.append(medicine_dict)
    return
```

```python
medicine_list = []
for i in range(1,21):
    table1(i)

final_table= pd.DataFrame(medicine_list)
```
